## Task 1: Describe Linear Regression in English
Given a dataset of $n$ values $(x_i, y_i)$ for $1 \leq i \leq n$, describe the process to compute values $\alpha$ and $\beta$ such that $$\sum_{i = 1}^n (y_i - \alpha x_i - \beta)^2$$ is minimal. This will yield a (linear) model $y = \alpha x + \beta$ adhering to the least-squares condition.
> Marking information: Up to 30 points: 15 points for an algorithm that can be followed, and 15 points for the correctness of this algorithm against the specification "linear regression." 


### Notes

My notes. The above formula is same as <br>$$\sum_{i = 1}^n (y_i -( \alpha x_i + \beta))^2$$$$=$$ $$\large (y - \hat y)^2$$<br>
where $y$ = actual value of y
    <br>&emsp; $\hat y$ = predicted value of y = $\alpha x + \beta$


Let us imagine there is a dataset $(x_i,y_i)$, where $x_i$ are input values and $y_i$ are output values. Here $y_i$ is dependant on value of $x_i$ but $x_i$ is an independent variable. $\alpha$ is the slope of straight line and $\beta$ is the y-intercept.

The equation of a plane in 2-dimension is $y$ = $\alpha$x + $\beta$. 

From the above equation we can observe that, there is a +ve slope for a line from bottom left to top right, and -ve slope if line is kept top left to bottom right. For every one unit increase in x the predicted value of y increases by the value of the slope. The y intercept is the location on the y axis where the line passes through. This is the value of y when x equals 0.

We have understood what the equation of line represents. Now, what we need to find is the values of $\alpha$ and $\beta$ for which, the sum of squared errors is the lowest. This line is the one that fits the given data most accurately.

So now, the optimization task is 

$$ min \sum_{i = 1}^n (y_i -( \alpha x_i + \beta))^2$$

As the above equation is minizmized, this means that the $\alpha$ and $\beta$ values for the equation of $y = \hat \alpha x + \hat \beta$ are optimized. <br> This means, for a given input of $x$, we can now get the optimal predicted value of $y$

## Implementing linear regression using simple linear regression method as in wikipedia

```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv('inp.csv')

#plotting data just to see what it looks like
'''plt.scatter(df.iloc[:,0], df.iloc[:,1])
plt.title('Scatter plot of data\n')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()'''

y_mean = df['GPA'].mean()
x_mean = df['SAT'].mean()

#initializing the values of numerator and denominator of alpha because we will need to summate it all in the for loop
top_sum = 0
low_sum = 0

#for each value in dataframe, get the summation values needed
for i in range(len(df)):
    y = df.iloc[i].GPA
    x = df.iloc[i].SAT
    top_sum += (x-x_mean)*(y-y_mean)
    low_sum += (x-x_mean)**2
    
#get the alpha and beta values
alpha = top_sum/low_sum
beta = y_mean - alpha*x_mean

#for prediction
x_1 = df.SAT
y_1 = alpha*x_1+beta
y_2 = df.GPA

#plotting data and regressor
plt.xlabel('SAT score')
plt.ylabel('GPA')
plt.title('SAT vs GPA of student')
plt.scatter(x_1, y_2, label='Student')
plt.plot(x_1,y_1,color='red',label='regressor')
plt.legend(loc='upper left')
plt.show()

print('Alpha value = ',round(alpha,3),'\nBeta value = ',round(beta,3), '\nEquation of the line: y =', round(alpha,3), 'x +',round(beta,3))
```

## Gradient Descent Optimized implementation of Linear Regression

Anology for GD:

(https://en.wikipedia.org/wiki/Gradient_descent#An_analogy_for_understanding_gradient_descent)

The basic intuition behind gradient descent can be illustrated by a hypothetical scenario. A person is stuck in the mountains and is trying to get down (i.e., trying to find the global minimum). There is heavy fog such that visibility is extremely low. Therefore, the path down the mountain is not visible, so they must use local information to find the minimum. They can use the method of gradient descent, which involves looking at the steepness of the hill at their current position, then proceeding in the direction with the steepest descent (i.e., downhill). If they were trying to find the top of the mountain (i.e., the maximum), then they would proceed in the direction of steepest ascent (i.e., uphill). Using this method, they would eventually find their way down the mountain or possibly get stuck in some hole (i.e., local minimum or saddle point), like a mountain lake. However, assume also that the steepness of the hill is not immediately obvious with simple observation, but rather it requires a sophisticated instrument to measure, which the person happens to have at the moment. It takes quite some time to measure the steepness of the hill with the instrument, thus they should minimize their use of the instrument if they wanted to get down the mountain before sunset. The difficulty then is choosing the frequency at which they should measure the steepness of the hill so not to go off track.

In this analogy, the person represents the algorithm, and the path taken down the mountain represents the sequence of parameter settings that the algorithm will explore. The steepness of the hill represents the slope of the function at that point. The instrument used to measure steepness is differentiation. The direction they choose to travel in aligns with the gradient of the function at that point. The amount of time they travel before taking another measurement is the step size.

Read https://en.wikipedia.org/wiki/Gradient_descent to understand the working of optimization.
